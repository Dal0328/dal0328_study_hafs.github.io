```python
import sys
import traceback
import tellopy
import av
import cv2  
import numpy
from PIL import Image
import os, glob, numpy as np
from keras.models import load_model

def main():
    drone = tellopy.Tello()
    image_w = 60
    image_h = 60
    caltech_dir = "C:/Users/dahyun/Desktop/ARC_AutoDriving_Drone/multi_img_data/imgs_others_test"
    pixels = image_h * image_w * 3
    model = load_model('C:/Users/dahyun/Desktop/ARC_AutoDriving_Drone/model/multi_img_classification.model')
    drone_drive = True
    try:
        drone.connect()
        drone.wait_for_connection(60.0)
        container = av.open(drone.get_video_stream())
        # skip first 300 frames
        frame_skip = 100
        drone.land()
        while drone_drive:
            for frame in container.decode(video=0): # 조정 필요
                if 0 < frame_skip:
                    frame_skip = frame_skip - 1
                    continue
                
                image = cv2.cvtColor(numpy.array(frame.to_image()), cv2.COLOR_RGB2BGR)
                cv2.imshow('Original', image)
                image_pil = Image.fromarray(image)
                image_pil.convert("LA")
                img = image_pil.resize((image_w, image_h))
                X = np.array([np.asarray(img)])
                prediction = model.predict(X)
                np.set_printoptions(formatter={'float': lambda x: "{0:0.3f}".format(x)})
                pre_ans = prediction[0].argmax()
                pre_ans_str = ''
                if pre_ans == 0: pre_ans_str = "a0"
                elif pre_ans == 1: pre_ans_str = "a1"
                elif pre_ans == 2: pre_ans_str = "b0"
                elif pre_ans == 3: pre_ans_str = "b1"
                elif pre_ans == 4: pre_ans_str = "c0"
                elif pre_ans == 5: pre_ans_str = "c1"
                if prediction[0][0] >= 0.8 : 
                    print(pre_ans_str+"으로 추정됩니다.")
                    drone.forward(1)
                if prediction[0][1] >= 0.8: 
                    print(pre_ans_str+"으로 추정됩니다.")
                    drone.clockwise(1)
                if prediction[0][2] >= 0.8: 
                    print(pre_ans_str+"으로 추정됩니다.")
                    drone.forward(1)
                if prediction[0][3] >= 0.8: 
                    print(pre_ans_str+"으로 추정됩니다.")
                    drone.clockwise(1)
                if prediction[0][4] >= 0.8: 
                    print(pre_ans_str+"으로 추정됩니다.")
                    drone.forward(1)
                if prediction[0][5] >= 0.8: 
                    print(pre_ans_str+"으로 추정됩니다.")
                    drone.land()
                cv2.waitKey(1)
                frame_skip = 30

    except Exception as ex:
        exc_type, exc_value, exc_traceback = sys.exc_info()
        traceback.print_exception(exc_type, exc_value, exc_traceback)
        print(ex)
    finally:
        drone.quit()
        cv2.destroyAllWindows()

main()
```
