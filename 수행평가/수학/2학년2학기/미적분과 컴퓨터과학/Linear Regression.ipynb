<h3> 초기값 설정 </h1>

```python
x_data = [1,2,3,4,5]
y_data = [-10,-5,3,6,9]

epoch = 1000
learning_rate = 0.01
```

<h3> 미분을 위한 sympy import 및 함수 설정 </h1>

```python
from sympy import symbols, Derivative
w,x,b= symbols('w,x,b')
fx = w*x+b
MeanSquareError = 0

for i in range(len(x_data)) :
    MeanSquareError += (y_data[i]-fx.subs({x:x_data[i]}))**2

MeanSquareError_w = Derivative(MeanSquareError, w).doit()
MeanSquareError_b = Derivative(MeanSquareError, b).doit()

print(MeanSquareError)
print(MeanSquareError_w)
print(MeanSquareError_b)
```

<h3> 학습 </h3>

```python
w_ = 1
b_ = 0
data = []

for i in range(epoch):
    w_ = w_ - learning_rate* (MeanSquareError_w.subs({w:w_,b:b_}))
    b_ = b_ - learning_rate* (MeanSquareError_b.subs({w:w_,b:b_}))
    data.append([w_,b_,MeanSquareError.subs({w:w_,b:b_})])
    
    print("="*40)
    print(w_,b_)
    print(MeanSquareError.subs({w:w_,b:b_}))
    print("="*40)
    print()
```

<h3> MeanSquareError 그래프 </h3>

```python
import matplotlib.pyplot as plt
plt.plot(range(1,len(data)+1),[i[2] for i in data])
plt.show()
```
